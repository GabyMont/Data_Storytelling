```python
# Visualizacion de correlacion de datos #
plt.rcParams['figure.figsize'] = (30, 30)
sns.heatmap(df_train.corr(numeric_only=True), annot = True, cmap = 'viridis')
plt.show()
```

```python
# Eliminación de variables con baja correlacion segun lo observado en el gráfico de calor #
Variable_elimin_train=['Id','MSSubClass','OverallCond','BsmtFinSF2','LowQualFinSF','BsmtHalfBath','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold','YrSold']
train_x=df_train.drop(labels=Variable_elimin_train, axis=1)
train_x.columns
```

```python
# Se vuelve a visualizar la correlación de datos #
plt.figure(figsize=(30,30))
sns.heatmap(train_x.corr(numeric_only=True), annot = True, cmap = 'viridis')
plt.show()
```

```python
# Se evalua los valores atipicos considerados mediante el gráfico de calor#
plt.figure(figsize=(10,10))
base_color = "#5A9"
dark_palette = sns.dark_palette(base_color, n_colors=10)
ax=sns.scatterplot(x=train_x.GrLivArea , y=train_x.SalePrice , hue=train_x.OverallQual , size=train_x.BedroomAbvGr, palette=dark_palette)
plt.title('Ground Living Area')
plt.show()
```

```python
plt.figure(figsize=(10,10))
base_color = "#5A9"
dark_palette = sns.dark_palette(base_color, n_colors=10)
ax=sns.scatterplot(x=train_x.LotArea , y=train_x.SalePrice , hue=train_x.OverallQual , size=train_x.BedroomAbvGr, palette=dark_palette)
plt.title('Lot Area')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x=train_x.SalePrice, color='lightseagreen')
```

```python
plt.figure(figsize=(20,20))
ax=sns.barplot(x="YearBuilt", y='SalePrice', data=train_x , errorbar=None ,color='lightseagreen')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.title('Evaluacion de Precio vs Habitaciones')
plt.show()
```
